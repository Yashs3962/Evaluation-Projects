```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

```

```python
url = "https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Temperature%20Forecast/temperature.csv"

try:
    temperature_data = pd.read_csv(url)
    print("Dataset loaded successfully.")
except Exception as e:
    print(f"Error loading dataset: {str(e)}")
```

```python
temperature_data.head()
temperature_data.info()
```

```python
temperature_data.dropna(inplace=True)
```

```python
X = temperature_data.drop(['Next_Tmax', 'Next_Tmin', 'Date'], axis=1)
y_max = temperature_data['Next_Tmax']
y_min = temperature_data['Next_Tmin']

```

```python
X_train, X_test, y_max_train, y_max_test = train_test_split(X, y_max, test_size=0.2, random_state=42)
X_train, X_test, y_min_train, y_min_test = train_test_split(X, y_min, test_size=0.2, random_state=42)
```

```python
rf_max = RandomForestRegressor(random_state=42)
rf_max.fit(X_train, y_max_train)
```

```python
y_max_pred = rf_max.predict(X_test)
```

```python
rf_min = RandomForestRegressor(random_state=42)
rf_min.fit(X_train, y_min_train)
```

```python
from sklearn.metrics import mean_squared_error
rmse_max = mean_squared_error(y_max_test, y_max_pred, squared=False)
print("Root Mean Squared Error (RMSE) for Next-Day Maximum Temperature:", rmse_max)
rmse_min = mean_squared_error(y_min_test, y_min_pred, squared=False)
print("Root Mean Squared Error (RMSE) for Next-Day Minimum Temperature:", rmse_min)
```
